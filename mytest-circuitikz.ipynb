```python
from tikz import *
import numpy as np
```

```python
x = np.linspace(-np.pi,np.pi,100)
y = np.sin(x)
sin_plot = list(zip(x,y))
z = np.exp(-x**2)
gauss_plot = list(zip(x,z))
```

```python
pic = Picture()
pic.draw(line(sin_plot),thick=True,color='red')
pic.draw(line(gauss_plot),thick=True)
pic.node(r'$\exp(-x^2)$','eq1',at=(0,1),anchor='south')
# pic.node()
pic.node(r'$\sin x$','eq2',at=(np.pi/2,1),anchor='south')
pic
```

```python
pic2 = Picture(scale=3)
six=[]
r=1
num=6
init=360/num/2
initcopy=init
for i in range(num):
    six.append(f'({initcopy}:{r})')
    initcopy+=360/num
six.append(f'({init}:{r})')
pic2.draw(line(six), thick=True)
for i in range(num):
    pic2.node(chr(ord('A') + i),chr(ord('A') + i),at=(six[i][:-2]+'1.1)'))

pic2.demo()
pic2.write_image('six.pdf')
```

```python
#符合电路的实验步骤示意图
pic3=Picture()
# pic3.style('every node', font=r'\tiny')
pic3.node('Pulse Signal 1','PS1',(0,0),draw=True,anchor='west')
pic3.draw('(PS1.east)',lineto('+(0.5,0)'),coordinate(name='PS1R'))
pic3.draw('(PS1R)',lineto('+(0.5,0.5)',op='|-'),
          node('Forming Channel 1','FC1',draw=True,anchor='west'),
          opt='-latex')
pic3.draw('(PS1R)',lineto('+(0.5,-0.5)',op='|-'),
          node('Forming Channel 2','FC2',draw=True,anchor='west'),
          opt='-latex')
pic3.node('Pulse Signal 2','PS2',(0,-1.5),draw=True,anchor='west',opt='-latex')
pic3.draw('(PS2.east)',lineto('+(1,0)'),
          node('Forming Channel 3','FC3',draw=True,anchor='west'),
          opt='-latex')
pic3.draw('(FC2.east)',lineto('+(0.5,0)'),coordinate(name='CLL'))
pic3.draw('(FC1.east)',lineto('(CLL)',op='-|'))
pic3.draw('(FC3.east)',lineto('(CLL)',op='-|'))
pic3.draw('(CLL)',lineto('+(0.5,0)'),
          node('Coincidence Logic Gate','CL',draw=True,anchor='west'),
          opt='-latex')
pic3.draw('(CL.east)',lineto('+(1,0)'),
          node('Counter','Counter',draw=True,anchor='west'),
          opt='-latex')

pic3.write_image('Schematic-diagram-of-the-measurement-experiment-circuit-of-electronics.pdf')
print(pic3.code())
pic3
```
